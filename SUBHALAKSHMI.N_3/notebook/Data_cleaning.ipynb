# **Data Cleaning and Exploratory Data Analysis (EDA)**

# **Introduction**

This project focuses on cleaning and analyzing a real-world dataset (Airbnb NYC 2019).
Data cleaning ensures the dataset is accurate, consistent, and usable for analysis.

The goal of this project is to:
1. Clean the dataset
2. Handle missing values and duplicates
3. Perform exploratory data analysis (EDA)
4. Visualize insights using interactive graphs

**Tools Used:**
- Python
- Pandas
- NumPy
- Plotly (for interactive visualization)


**Import Libraries**

```python
import pandas as pd
import numpy as np
import plotly.express as px
```

**Load Dataset**

```python
df = pd.read_csv('AB_NYC_2019.csv')

# Display first 5 rows
df.head()
```

** Dataset Overview**

```python
df.info()

df.describe()
```

# Data Cleaning

**Description:**

Data cleaning is the process of fixing or removing incorrect, corrupted,
duplicate, or incomplete data.

**Key Tasks:**
- Handle missing values
- Remove duplicates
- Ensure consistency


```python


# Check missing values
df.isnull().sum()
```

 **Missing Data Handling**

```python
# Fill missing 'reviews_per_month' with 0
df['reviews_per_month'].fillna(0, inplace=True)

# Drop unnecessary columns with too many nulls
df.drop(['last_review'], axis=1, inplace=True)

df.isnull().sum()
```

**Duplicate Removal**

```python
df.drop_duplicates(inplace=True)

df.shape
```

```python
# Remove listings with price > 500
df = df[df['price'] < 500]
```

# Pie Chart

```python
fig = px.pie(df, names='room_type', title='Room Type Distribution')
fig.show()
```


# Scatter Plot



Description:
Shows relationship between price and number of reviews.
Hover to see details.


```python


fig = px.scatter(
    df,
    x='number_of_reviews',
    y='price',
    color='room_type',
    title='Price vs Reviews'
)

fig.show()
```


# Heatmap

Description:
Shows correlation between numerical variables.


```python


corr = df.corr(numeric_only=True)

fig = px.imshow(
    corr,
    text_auto=True,
    title='Correlation Heatmap'
)

fig.show()
```


# Bar Chart

Description:
Average price by neighbourhood group.



```python

avg_price = df.groupby('neighbourhood_group')['price'].mean().reset_index()

fig = px.bar(
    avg_price,
    x='neighbourhood_group',
    y='price',
    title='Average Price by Area'
)

fig.show()
```

# Conclusion / Result


**Conclusion:**

The data cleaning process improved dataset quality by handling missing values,
removing duplicates, and filtering outliers.

**Key Insights:**
- Most listings are Private Rooms.
- Manhattan has higher average prices.
- Listings with more reviews tend to have moderate pricing.
- Strong correlations exist between reviews and availability.

**Result:**
This analysis helps businesses and customers understand pricing trends,
popular room types, and location-based insights, enabling better decision-making.
